```python
import pandas as pd
import numpy as np
import copy
```

Превращаем цену в тип float64

```python
a = pd.read_excel("DataSet2021Spring_ОСНОВНОЙ.xlsx")
a = a.astype({'Price': 'float64'})
```

Соединяем цену со следующим столбцом, потому что цена -- число с плавающей запятой, а в России принято разделять запятые с помощью запятой. Значит, часть цены после запятой попала в следующий столбец. (Будем считать, что цена на данный товар АБСОЛЮТНО всегда имеет дробную часть, как и в настоящих магазинах (xx,99 копеек). 

```python
print(a.values[0][24])
for i in range(0, len(a.values)):
    temp = str(a.values[i][24])
    temp2 = str(a.values[i][25])
    if (temp != 'nan' and temp2 != 'None'):
        tmp = a.values[i]
        tmp[20] += tmp[21]*(10**(-len(str(tmp[21]))))
        tmp = np.delete(tmp, 21)
        a.loc[i] = np.append(tmp, None)
```

Ту же самую операцию проводим с Qty_Stock, предполагая, что на складе всегда остается больше товара, чем на полках в магазине. Количество на складе может быть дробным числом, потому что на складе часть могла порваться (если линолеум), разбиться, исчезнуть "в неизвестном направлении", поэтому на складе из полного числа вещей, грубо говоря, вычитается процент "потерь", зависящий от конкретной позиции

```python
for i in range(0, len(a.values)):
    temp = str(a.values[i][24])
    if (temp != 'nan' and temp != 'None'):
        tmp = a.values[i]
        tmp[21] += tmp[22]*(10**(-len(str(tmp[22]))))
        tmp = np.delete(tmp, 22)
        a.loc[i] = np.append(tmp, None)
```

Поскольку в датасете присутствуют ошибки (например, больше присутствует на складе, чем на полках, или отрицательное количество присутствия и того, и другого), то можно пойти двумя путями:
- Исправить их (путем свапа или смены знака
- Не трогать

И тот, и другой варианты приводят к изменению чистоты датасета, однако неизвестно, какой куда изменяет. Поэтому оставим оба датасета

```python
for i in range(0, len(a.values)):
    tmp = a.values[i]
    if abs(tmp[21]) < abs(tmp[22]):
        tmp[21], tmp[22] = tmp[22], tmp[21]
    if tmp[21] < 0:
        tmp[21] = -tmp[21]
    if tmp[22] < 0:
        tmp[22] = -tmp[22]
    tmp[22] = int(tmp[22])
    a.loc[i] = tmp
```
